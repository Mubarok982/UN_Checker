```python
from bs4 import BeautifulSoup

# Fungsi ekstraksi username dari HTML
def extract_usernames_from_html(file_path):
    with open(file_path, "r", encoding="utf-8") as f:
        soup = BeautifulSoup(f, "html.parser")
        links = soup.find_all("a", href=True)
        usernames = [link["href"].replace("https://www.instagram.com/", "").strip("/") for link in links]
        return set(usernames)

# Ganti path ini sesuai file kamu
followers = extract_usernames_from_html("followers_1.html")
following = extract_usernames_from_html("following.html")

# Akun yang tidak follow balik
not_following_back = sorted(following - followers)

# Tampilkan tabel
print("| No | Username | Link |")
print("|----|----------|------|")
for i, username in enumerate(not_following_back, 1):
    print(f"| {i} | {username} | https://instagram.com/{username} |")

print(f"\nTotal akun yang tidak follow balik: {len(not_following_back)}")
```
