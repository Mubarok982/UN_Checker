```python
from bs4 import BeautifulSoup
import re

def extract_usernames_from_html(file_path):
    usernames = set()

    with open(file_path, "r", encoding="utf-8") as f:
        html = f.read()

    soup = BeautifulSoup(html, "html.parser")


    for a in soup.find_all("a", href=True):
        href = a["href"]

        match = re.search(r"instagram\.com/_u/([^/?\"']+)", href)
        if match:
            usernames.add(match.group(1))

        match2 = re.search(r"instagram\.com/([^/?\"']+)", href)
        if match2:
            usernames.add(match2.group(1))

    for text in soup.stripped_strings:
        if re.fullmatch(r"[A-Za-z0-9._]+", text):
            usernames.add(text)

    return usernames
```

```python
followers = extract_usernames_from_html("followers_1.html")
following = extract_usernames_from_html("following.html")

not_follow_back = sorted(following - followers)

print("| No | Username | Link |")
print("|----|----------|------|")

for i, u in enumerate(not_follow_back, start=1):
    print(f"| {i} | {u} | https://instagram.com/{u} |")

print(f"\nTotal akun yang tidak follow balik: {len(not_follow_back)}")
```
